```python
import time
import cv2
import requests
import numpy as np
from google.colab.patches import cv2_imshow

# Function to draw bounding boxes on the image
def draw_boxes(image, predictions):
    for prediction in predictions:
        x, y, w, h = int(prediction['x']), int(prediction['y']), int(prediction['width']), int(prediction['height'])
        confidence = prediction['confidence']
        class_name = prediction['class']
        color = (0, 255, 0)  # Green color for the bounding box
        cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)
        text = f'{class_name} {confidence:.2f}'
        cv2.putText(image, text, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

# import the inference-sdk
from inference_sdk import InferenceHTTPClient

# initialize the client
CLIENT = InferenceHTTPClient(
    api_url="https://detect.roboflow.com",
    api_key="JZWkVD14Vla1TbQHIAEg"
)

# infer on a local image
result = CLIENT.infer("aa.jpeg", model_id="breast-cancer-hsjck/6")

# Adjust the confidence threshold
confidence_threshold = 0.5  # Set your desired confidence threshold (e.g., 0.5 for 50% confidence)
filtered_predictions = [pred for pred in result['predictions'] if pred['confidence'] >= confidence_threshold]

# Implement logic to handle overlapping bounding boxes
def remove_overlapping_boxes(predictions, overlap_threshold):
    # Sort predictions by confidence score (highest to lowest)
    predictions.sort(key=lambda x: x['confidence'], reverse=True)
    # Initialize list to store non-overlapping predictions
    non_overlapping_predictions = []
    # Iterate through predictions
    for pred in predictions:
        # Check if current prediction overlaps with any existing non-overlapping prediction
        overlap = False
        for existing_pred in non_overlapping_predictions:
            if calculate_overlap(pred, existing_pred) > overlap_threshold:
                overlap = True
                break
        # If no overlap, add current prediction to non-overlapping predictions
        if not overlap:
            non_overlapping_predictions.append(pred)
    return non_overlapping_predictions

# Function to calculate overlap between two bounding boxes
def calculate_overlap(box1, box2):
    x1 = max(box1['x'], box2['x'])
    y1 = max(box1['y'], box2['y'])
    x2 = min(box1['x'] + box1['width'], box2['x'] + box2['width'])
    y2 = min(box1['y'] + box1['height'], box2['y'] + box2['height'])
    intersection = max(0, x2 - x1) * max(0, y2 - y1)
    area1 = box1['width'] * box1['height']
    area2 = box2['width'] * box2['height']
    union = area1 + area2 - intersection
    overlap_ratio = intersection / union
    return overlap_ratio

# Remove overlapping boxes
non_overlapping_predictions = remove_overlapping_boxes(filtered_predictions, overlap_threshold=0.3)

# Load the image
image = cv2.imread("aa.jpeg")

# Draw non-overlapping bounding boxes on the image
draw_boxes(image, non_overlapping_predictions)

# Display the image with predictions
cv2_imshow(image)
time.sleep(0.1)
```
